```
stellar_objects = DataBlock(blocks=(ImageBlock, CategoryBlock), 
                            get_items=get_image_files,
                            splitter=RandomSplitter(seed=42),
                            get_y=parent_label,
                            item_tfms=Resize(128)
                            )
```

```
dls = stellar_objects.dataloaders(path)
```

```
dls.valid.show_batch(max_n=8)
```

```
stellar_objects = stellar_objects.new(item_tfms=Resize(224, ResizeMethod.Pad, pad_mode='zeros'), batch_tfms=aug_transforms())
dls = stellar_objects.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(4)
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_top_losses(6, nrows=3)
```

```
cleaner = ImageClassifierCleaner(learn)
```

```
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
```

```
for idx, cat in cleaner.change(): 
  curname = str(cleaner.fns[idx])
  newname = str(idx+20)+'.jpg'
  print(path/cat/newname)
  os.rename(curname, path/cat/newname)
  
```

```
path_new = Path()
path_new.ls(file_exts='.pkl')
```

```
btn_upload = widgets.FileUpload()
```

```
new_learn = load_learner(path_new/'export.pkl')
```

```
out_pl = widgets.Output()
```

```
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```
btn_run = widgets.Button(description='Classify')
```

```
def on_click_classify(change):
  img = PILImage.create(btn_upload.data[-1])
  out_pl.clear_output()
  with out_pl: display(img.to_thumb(224, 224))
  pred, pred_idx, probs = new_learn.predict(img)
  lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```
VBox([widgets.Label('Upload an object'),
      btn_upload, btn_run, out_pl, lbl_pred])
```
